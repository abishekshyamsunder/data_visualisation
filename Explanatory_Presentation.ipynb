## Investigation Overview

> This presentation aims to provide insights into the dataset described below visually. It aims to answer the questions.
1. What is the distribution of rides across the day?
2. Are there any differences between the rides of Customers and Subscribers?
3. Are there any differences between the rides across the week/ between weekdays and weekends?

## Dataset Overview

> The dataset used here is provided by Bay Area Motivate, LLC ("Motivate"), and it contains bike ride data collected between June 2017 and May 2018  
> The dataset is contains 1,142,896 entries and 20 columns  

```python
# import all packages and set plots to be embedded inline
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sb
import geopy.distance
from datetime import datetime
%matplotlib inline

# suppress warnings from final output
import warnings
warnings.simplefilter("ignore")
```

```python
# load in the dataset into a pandas dataframe
df = pd.read_csv('bikedatafull.csv')
```

## (Visualization 1)

> Visualization1 helps us understand the distribution of rides across the day. (i.e. a count was taken for each hour of the day), with the subslide showing the distribution for each user type individually  
> SubSlide 1: The number of rides peaks at 8 in the morning and at 5 in the evening, especially due to Subscribers

```python
base_color = sb.color_palette()[2]
sb.catplot(x="hour", hue="user_type", kind="count",palette="pastel", edgecolor=".6",data=df);
plt.title('Distribution of rides across the day');
```

## (Visualization 2)

> Visualization 2 helps us understand the differences between the trips of Consumers and Subscribers  
> SubSlide 1: Despite the average distance travelled each month by Subscribers and Customers being almost equal, the average duration of trips for Subscribers is much lesser across the year

```python
f, axes = plt.subplots(1, 2, figsize=(10, 5), sharex=True)
sb.despine(left=True)

ax = sb.barplot(data = df, x = 'month', y = 'duration_sec', hue="user_type",ax=axes[0]);
plt.ylabel('average duration');
ax.legend_.remove()
ax.set_title('Average Duration Each month')
ax.set_ylabel('Average Duration (seconds)')

ax2 = sb.barplot(data = df, x = 'month', y = 'distance', hue="user_type",ax=axes[1]);
plt.ylabel('average distance (kms)');
ax2.set_title('Average Distance Each month')
plt.legend(bbox_to_anchor=(1.05, 1), loc=2, borderaxespad=0.)

```

## (Visualization 3)

> Helps us understand if there are any differences between the rides in weekdays and weekends  
> SubSlide 1: Despite the average distance being slightly lesser in weekends, the average duration for each trip is much higher  
> SubSlide 2: This trend is seen in both Customers and Subscribers  
> Subslide 3: Total Number of rides in weekends and weekdays compared. Number of rides on weekends is lesser


```python
base_color = sb.color_palette()[0]
f, axes = plt.subplots(1, 2, figsize=(15, 5), sharex=True)
sb.despine(left=True)

ax = sb.barplot(data = df, x = 'day', y = 'duration_sec',color = base_color,ax=axes[0]);
ax.set_ylabel('Average Duration (seconds)')
ax.set_title('Average Duration each day')

ax2 = sb.barplot(data = df, x = 'day', y = 'distance',color = base_color,ax=axes[1]);
plt.ylabel('average distance (kms)');
ax2.set_title('Average Distance each day')
```

```python
base_color = sb.color_palette()[0]
fig = plt.figure(figsize=(10,5))
sb.barplot(data = df, x = 'day', y = 'duration_sec', hue="user_type");
plt.ylabel('average duration (seconds)');
plt.legend(bbox_to_anchor=(1.05, 1), loc=2, borderaxespad=0.)
plt.title('Average Duration of trips across the week')
```
